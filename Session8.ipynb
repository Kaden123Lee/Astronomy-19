```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array x from 0 to 1 with 100 steps
x = np.linspace(0, 1, 100)

# Define functions for sin(x) and cos(x)
def sin_x(x):
    return np.sin(2 * np.pi * x)

def cos_x(x):
    return np.cos(2 * np.pi * x)
```

```python
# Create a figure with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot sin(x) vs. x in the left panel
axs[0].plot(x, sin_x(x))
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].set_title('sin(x)')

# Plot cos(x) vs. x in the right panel
axs[1].plot(x, cos_x(x))
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].set_title('cos(x)')

# Adjust layout and save the figure as a PDF
plt.tight_layout()
plt.savefig('sin_cos_plot.pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
def exp_func(x):
    return np.exp(x)
```

```python
y = exp_func(x)
plt.plot(x, y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Exponential Growth of Awesomeness')
plt.savefig('exponential_growth.pdf')
plt.show()
```
